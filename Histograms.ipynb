```python
import seaborn as sns
sns.set(style="darkgrid")

tips = sns.load_dataset("tips")
sns.jointplot("total_bill", "tip", data=tips,
                  kind="reg", truncate=False,
                  xlim=(0, 60), ylim=(0, 12),
                  color="b", height=10)
```
